```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 15 11:01:32 2022

@author: Simone
"""

import numpy as np 
import pandas as pd 
#from random import randrange
#import random
#import sklearn
df = pd.read_csv("new/popolazioneEtaSesso2021.csv")
df
```

```python
latLong = pd.read_csv("new/italy_lng_latAgg2021.csv")
latLong
```

```python
dfNoAreas = df.loc[df['ITTER107'].str.isdigit()]
dfAreas = df.loc[df['ITTER107'].str.isdigit()==False]
#dfNoAreas['ITTER107'] = dfNoAreas['ITTER107'].astype('int64')
dfNoAreas = dfNoAreas.astype({"ITTER107": int})
dfNoAreas
```

```python
dfAreas
```

```python
tot = pd.merge(dfNoAreas, latLong,how='outer',left_on=['ITTER107'],right_on=['istat'],indicator=True)
#tot = tot.astype({"ITTER107": int, "istat":int})
tot
```

```python
# tot['ITTER107'].isnull().sum()
# tot.loc[tot['_merge'] == "right_only"]
```

```python
# tot.loc[(tot['_merge'] == "left_only")]
# #tot.loc[tot['ITTER107'] == 1317]
```

```python
tot = tot.drop(['istat','comune','_merge'], axis=1)
tot
```

```python
allLatLong = pd.concat([tot,dfAreas], ignore_index=True)
allLatLong.to_csv("popSessoEtaLngLat2021.csv", encoding='utf-8',index=False, na_rep="")
allLatLong
```
