## Preprocessing marital status

```python
mar =  pd.read_csv("new/maritalStatus2020.csv", dtype = {"ITTER107":str})
mar['ITTER107'] = mar['ITTER107'].str.lstrip('0')
mar
```

#### Fix problematic municipalities
Ville di Fiemme, Novella, Borgo d'Anaunia, Faedo in San Michele all'Adige, Vendrogno in Bellano, Monteciccardo in Pesaro

```python
#Novella
novella = mar.loc[(mar['Territorio'] == "Cagnò") | (mar['Territorio'] == "Brez") | 
        (mar['Territorio'] == "Cloz") | (mar['Territorio'] == "Revò") |
        (mar['Territorio'] == "Romallo")]
mar = mar.drop(novella.index)#Remove from dataset
group = novella.groupby(['Sesso','Età','Stato civile'], as_index=False).agg({'Value':'sum'})
group['ITTER107'] = "22253"
group['Territorio'] = "Novella"
mar = pd.concat([mar, group], ignore_index=True)

#Borgo d'Anaunia
borgo = mar.loc[(mar['Territorio'] == "Castelfondo") | (mar['Territorio'] == "Fondo") | 
        (mar['Territorio'] == "Malosco")]
mar = mar.drop(borgo.index)#Remove from dataset
group = borgo.groupby(['Sesso','Età','Stato civile'], as_index=False).agg({'Value':'sum'})
group['ITTER107'] = "22252"
group['Territorio'] = "Borgo d'Anaunia"
mar = pd.concat([mar, group], ignore_index=True)

#Ville di Fiemme
ville = mar.loc[(mar['Territorio'] == "Carano") | (mar['Territorio'] == "Daiano") | 
        (mar['Territorio'] == "Varena")]
mar = mar.drop(ville.index)#Remove from dataset
group = ville.groupby(['Sesso','Età','Stato civile'], as_index=False).agg({'Value':'sum'})
group['ITTER107'] = "22254"
group['Territorio'] = "Ville di Fiemme"
mar = pd.concat([mar, group], ignore_index=True)

#San Michele all'Adige
sm = mar.loc[(mar['Territorio'] == "San Michele all'Adige") | 
        (mar['Territorio'] == "Faedo")]
mar = mar.drop(sm.index)#Remove from dataset
group = sm.groupby(['Sesso','Età','Stato civile'], as_index=False).agg({'Value':'sum'})
group['ITTER107'] = "22167"
group['Territorio'] = "San Michele all'Adige"
mar = pd.concat([mar, group], ignore_index=True)
        
#Bellano
bell = mar.loc[(mar['Territorio'] == "Vendrogno") | 
        (mar['Territorio'] == "Bellano")]
mar = mar.drop(bell.index)#Remove from dataset
group = bell.groupby(['Sesso','Età','Stato civile'], as_index=False).agg({'Value':'sum'})
group['ITTER107'] = "97008"
group['Territorio'] = "Bellano"
mar = pd.concat([mar, group], ignore_index=True)
        
#Pesaro
pes = mar.loc[(mar['Territorio'] == "Pesaro") | 
        (mar['Territorio'] == "Monteciccardo")]
mar = mar.drop(pes.index)#Remove from dataset
group = pes.groupby(['Sesso','Età','Stato civile'], as_index=False).agg({'Value':'sum'})
group['ITTER107'] = "41044"
group['Territorio'] = "Pesaro"
mar = pd.concat([mar, group], ignore_index=True)
```

```python
mar = mar.loc[(mar['Stato civile'] != 'unito/a civilmente ') & 
             (mar['Stato civile'] != 'già in unione civile (per decesso del partner) ') &
             (mar['Stato civile'] !=  'già in unione civile (per scioglimento unione)')]
```

```python
mar.to_csv("maritalStatus2020Agg2021.csv", encoding='utf-8',index=False)
```
